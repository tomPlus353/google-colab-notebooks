```python
import pandas as pd
import numpy as np

def analyze_trade_price_variation(df: pd.DataFrame, column_name: str):
    """
    Analyzes the variation of 'TradePrice' across different categories
    within a specified string column of a DataFrame.

    Args:
        df (pd.DataFrame): The input DataFrame.
        column_name (str): The name of the categorical string column to analyze.

    Returns:
        None: Prints the analysis results.
    """
    print(f"Analyzing TradePrice variation by '{column_name}':\n")

    # Check if necessary columns exist
    if column_name not in df.columns:
        print(f"Error: Column '{column_name}' not found in the DataFrame.")
        return
    if 'TradePrice' not in df.columns:
        print("Error: 'TradePrice' column not found in the DataFrame.")
        return

    # Ensure TradePrice is numeric and handle non-numeric values by coercing to NaN
    df['TradePrice'] = pd.to_numeric(df['TradePrice'], errors='coerce')

    # Filter out rows with NaN in the target column or TradePrice
    cleaned_df = df.dropna(subset=[column_name, 'TradePrice']).copy()

    # Check if the column is of a string/object type after dropping NaNs
    if not pd.api.types.is_string_dtype(cleaned_df[column_name]):
        print(f"Warning: Column '{column_name}' is not of string type. Attempting to convert.")
        try:
            cleaned_df[column_name] = cleaned_df[column_name].astype(str)
        except Exception as e:
            print(f"Error: Could not convert '{column_name}' to string type: {e}")
            return

    if cleaned_df.empty:
        print("No valid data after cleaning for analysis.")
        return

    # Calculate mean TradePrice for each unique value in the specified column
    mean_prices = cleaned_df.groupby(column_name)['TradePrice'].mean().sort_values(ascending=False)

    # Calculate the variation (standard deviation of the mean prices)
    variation_std = mean_prices.std()
    price_range = mean_prices.max() - mean_prices.min()

    # print("Mean TradePrice per category:")
    # display(mean_prices)

    print(f"\nStandard Deviation of Mean TradePrices: {variation_std:,.2f}")
    print(f"Range of Mean TradePrices (Max - Min): {price_range:,.2f}")


    if variation_std == 0 or mean_prices.mean() == 0:
        cv = 0
    else:
        cv = (variation_std / mean_prices.mean()) * 100
    print(f"Coefficient of Variation (Std/Mean * 100): {cv:.2f}%")

    print("----------"*5)
    print("\n"*3)
    print("----------"*5)
```

```python
# District
analyze_trade_price_variation(data, 'DistrictName')

# stations
analyze_trade_price_variation(data, 'NearestStation')

# subset of stations
analyze_trade_price_variation(filtered_data, 'NearestStation')
```

```python
# Analyze variation by 'Type'
analyze_trade_price_variation(data, 'Type')

# Analyze variation by 'Region'
analyze_trade_price_variation(data, 'Region')

analyze_trade_price_variation(data, 'Municipality')
```

```python
# Analyze variation by 'Structure'
analyze_trade_price_variation(data, 'Structure')

# Analyze variation by 'CityPlanning'
analyze_trade_price_variation(data, 'CityPlanning')

# Analyze variation by 'Renovation'
analyze_trade_price_variation(data, 'Renovation')
```
